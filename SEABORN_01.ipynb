#LINE PLOT

```python
#line plot in seaborn.
var=[1,2,3,4,5,6,7]
var_1=[2,3,4,5,1,6,7]
x_1=pd.DataFrame({'var':var,'var_1':var_1})
sns.lineplot(x='var',y='var_1',data=x_1)
plt.show()
```

```python
y1=sns.load_dataset('penguins').head(20) #will only consider the data within first 20 rows.
y1
```

```python
sns.lineplot(x='bill_length_mm',y='bill_depth_mm',data=y1,hue='sex',style='sex',palette='Blues',markers=['o','^']) #hue will differ sexes,2 lines.
plt.grid() #squares,across the graph.
plt.title("WATASHINO LINE PLOT")
plt.show() #style will create dotted(Female),palette for color of graph line,markers to mark where graph changes rapidly.
```

# BAR PLOT

```python
var=sns.load_dataset("penguins")
var
```

```python
sns.barplot(x='island',y='bill_length_mm',data=var,hue='sex',palette='Accent') #here,we are calling col-data as string. 
plt.xlabel("Orange Islands")
plt.grid()
plt.ylabel("bill_lambai")
plt.title("BARPLOT_BAE")
plt.show()
```

```python
order_1=['Biscoe','Dream','Torgersen'] # creating a list of our desired order of Islands.
sns.barplot(x=var.island,y=var.bill_length_mm,hue=var.species,palette='Accent',order=order_1) #here,we are calling col-data as a subject of "var" dataframe.
plt.show() # with order parameter,you can change the order of x-axis's parameters.
```

```python
sns.set(style='darkgrid') #alternative of grid function for seaborn.
sns.barplot(x=var.island,y=var.bill_length_mm,hue=var.sex,palette='icefire',saturation=1,errcolor='b',capsize=0.3) 
plt.show() #saturation,modifies the saturation level(dark/light in color) of plots.
```

#HISTORAM PLOT

```python
sns.displot(var['bill_length_mm'])
plt.show()
```

```python
sns.displot(var['bill_length_mm'],bins=[30,35,40,45,50,55,60],kde=True,rug=True) #provides a regression line over your graph.
plt.show() #bins KW is used to space the bar plots. #a small line at the edge of the plot to show data distribution, it is dense where R.line is high.
```

```python
sns.displot(var['body_mass_g'],color='g')
plt.show()
```
